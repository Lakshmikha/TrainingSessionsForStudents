```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def estimate_coef(x, y):
    """
    Calculates the coefficients of the simple linear regression model.

    Args:
        x: Independent variable (features).
        y: Dependent variable (target).

    Returns:
        A tuple containing the intercept (b0) and slope (b1) of the regression line.
    """
    n = np.size(x)
    m_x, m_y = np.mean(x), np.mean(y)

    SS_xy = np.sum(y * x) - n * m_y * m_x          #sum of product of deviations by using a series formula
    SS_xx = np.sum(x * x) - n * m_x * m_x          #sum of squared deviations using series formula

    b_1 = SS_xy / SS_xx
    b_0 = m_y - b_1 * m_x

    return (b_0, b_1)

def plot_regression_line(x, y, b):
    """
    Plots the actual data points and the regression line.

    Args:
        x: Independent variable (features).
        y: Dependent variable (target).
        b: A tuple containing the intercept (b0) and slope (b1) of the regression line.
    """
    plt.scatter(x, y, color="r", marker="p", s=20) #s means size, marker is shape of point, scatter plotting, m-magenta

    y_pred = b[0] + b[1] * x

    plt.plot(x, y_pred, color="g")

    plt.xlabel('x')
    plt.ylabel('y')

    plt.show()

def main():

    data = pd.read_csv('hw.csv')


    x = data['x'].values
    y = data['y'].values

    # Estimate coefficients
    b = estimate_coef(x, y)
    print("Estimated coefficients:\nb_0 = {} \nb_1 = {}".format(b[0], b[1]))

    # Plot the regression line
    plot_regression_line(x, y, b)

if __name__ == "__main__":
    main()
```
