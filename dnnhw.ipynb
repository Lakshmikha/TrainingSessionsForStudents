```python
# filepath: [dnnhw.py](http://_vscodecontentref_/0)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
import numpy as np
from PIL import Image
import os

# 1. Define the model architecture
model = keras.Sequential([
    Dense(128, activation='relu', input_shape=(784,)),  # Input layer with 784 features
    Dense(1, activation='sigmoid')  # Output layer with 1 neuron for binary classification
])

# 2. Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # Binary classification loss
              metrics=['accuracy'])

# 3. Load and preprocess custom dataset
train_data_dir = '/content/drive/MyDrive/extracted_images_train'  # Replace with your training data path
test_data_dir = '/content/drive/MyDrive/extracted_images_test'  # Replace with your test data path

# Load training data
x_train = []
y_train = []
for label in os.listdir(train_data_dir):  # Assuming subdirectories are class labels
    label_dir = os.path.join(train_data_dir, label)
    if os.path.isdir(label_dir):
        for img_file in os.listdir(label_dir):
            img_path = os.path.join(label_dir, img_file)
            img = Image.open(img_path).convert('L')  # Convert to grayscale
            img = img.resize((28, 28))  # Resize to 28x28
            img_array = np.array(img).astype('float32') / 255.0  # Normalize
            x_train.append(img_array.flatten())  # Flatten to 1D array
            y_train.append(0 if label == 'cats' else 1)  # 0 for cats, 1 for dogs

x_train = np.array(x_train)
y_train = np.array(y_train)  # Labels as a numpy array

# Load test data
x_test = []
y_test = []
for label in os.listdir(test_data_dir):  # Assuming subdirectories are class labels
    label_dir = os.path.join(test_data_dir, label)
    if os.path.isdir(label_dir):
        for img_file in os.listdir(label_dir):
            img_path = os.path.join(label_dir, img_file)
            img = Image.open(img_path).convert('L')  # Convert to grayscale
            img = img.resize((28, 28))  # Resize to 28x28
            img_array = np.array(img).astype('float32') / 255.0  # Normalize
            x_test.append(img_array.flatten())  # Flatten to 1D array
            y_test.append(0 if label == 'cats' else 1)  # 0 for cats, 1 for dogs

x_test = np.array(x_test)
y_test = np.array(y_test)  # Labels as a numpy array

# 4. Train the model
model.fit(x_train, y_train, epochs=50, batch_size=32)

# 5. Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy)

# 6. Make predictions for a single test image
test_image_path = 'dog_28.jpg'  # Replace with your test image path
img = Image.open(test_image_path).convert('L')  # Convert to grayscale
img = img.resize((28, 28))  # Resize to 28x28
img_array = np.array(img).astype('float32') / 255.0  # Normalize
img_array = img_array.flatten().reshape(1, 784)  # Flatten and reshape for the model

predictions = model.predict(img_array)  # Predict probabilities
predicted_label = 1 if predictions[0][0] > 0.5 else 0  # Threshold at 0.5
print('Predicted label for the test image:', 'Dog' if predicted_label == 1 else 'Cat')
```
